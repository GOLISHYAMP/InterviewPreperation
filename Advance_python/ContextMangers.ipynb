# Context managers
A context manager in Python is a construct that allows you to manage resources efficiently, ensuring that setup and cleanup logic is handled properly. The most common use case is resource management, such as opening and closing files or database connections.

1. with Statement:

The with statement is used to wrap the execution of a block of code with methods defined by a context manager.
It ensures that resources are released, even if an error occurs

```python
with open('file1.txt', 'r') as f:
    print(f.read())
```

```python
class MyContextManager:
    def __enter__(self):
        print("enter in to the function")
        return self
    
    def __exit__(self, exc_type, exc_value, traceback):
        print("Exited the function")

with MyContextManager() as cm:
    print('data to print')
```
